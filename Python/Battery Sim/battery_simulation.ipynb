Basic Sim

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Parameters
capacity_ah = 3.0               # Battery capacity in Ah
discharge_current = 1.0         # Constant discharge current in A
runtime_hours = capacity_ah / discharge_current  # Time to full discharge in hours
runtime_seconds = int(runtime_hours * 3600)      # Convert to seconds
time = np.arange(0, runtime_seconds, 60)         # Time in seconds (step = 60s)

# Voltage model: simple linear drop
voltage_start = 4.2
voltage_end = 3.0
voltage = np.linspace(voltage_start, voltage_end, len(time))

# State of Charge (SoC)
soc = 100 * (1 - time / time[-1])  # From 100% to 0%

# Plot
plt.figure(figsize=(10, 5))
plt.subplot(2, 1, 1)
plt.plot(time / 60, voltage, label="Voltage (V)", color='tab:blue')
plt.ylabel("Voltage (V)")
plt.grid(True)
plt.legend()

plt.subplot(2, 1, 2)
plt.plot(time / 60, soc, label="SoC (%)", color='tab:green')
plt.xlabel("Time (minutes)")
plt.ylabel("State of Charge (%)")
plt.grid(True)
plt.legend()

plt.suptitle("Simple Battery Discharge Simulation")
plt.tight_layout()
plt.show()
```
